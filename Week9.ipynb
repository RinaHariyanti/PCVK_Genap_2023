### **Import Library**

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# ===============Dilasi Tanpa Library==============
img10 = cv2.imread('/content/drive/MyDrive/PCVK/Images/morphology.png',0)
#img5 = cv2.imread('/content/drive/MyDrive/PCVK/Images/fingerprint.png',0)

#Acquire size of the image
p, q= img10.shape

#Show the image
plt.imshow(img10, cmap="gray")

#Define new image to store the pixels of dilated image
imgDilate = np.zeros((p,q), dtype=np.uint8)

#Define the structuring element 
SED = np.array([[1,1,1], [1,1,1], [1,1,1]])
constant1 = 1

#Dilation operation without using inbuilt cv2 function
for i in range(constant1, p-constant1):
  for j in range(constant1,q-constant1):
    temp = img10[i-constant1:i+constant1+1, j-constant1:j+constant1+1]
    product = temp * SED
    imgDilate[i,j] = np.max(product)

plt.subplot(131),plt.imshow(img10, cmap="gray"),plt.title('Citra Awal')
plt.subplot(132),plt.imshow(imgDilate, cmap="gray"),plt.title('Citra Hasil Dilasi')
```

```python
# 3. Dilasi tanpa library
img = cv2.imread('/content/drive/MyDrive/PCVK/Images/morphology.png')

kernel = np.ones((3,3), np.uint8) # fungsi ones -> Membuat array dengan nilai 1, uint8 (unsigned integer 8 bit) -> tipe data yg digunakan untuk membuat gambar dengan menggunakan filter RGB dengan pixel dalam range [0,255].
dilation = cv2.dilate(img, kernel, iterations = 1) #iterator objek yang dapat diulangi

plt.subplot(121),plt.imshow(img),plt.title('Citra Awal') #fungsi subplots() kita bisa menggambar banyak plot dalam satu gambar.
plt.subplot(122),plt.imshow(dilation),plt.title('Citra Hasil Dilasi')
     
```

```python
# 4. Dilasi dengan library
kernel = np.ones((3,3), np.uint8)
dilation = cv2.morphologyEx(img, cv2.MORPH_DILATE, kernel) 

plt.subplot(121), plt.imshow(img), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(dilation), plt.title('Citra Hasil Dilasi')
```

```python
# ==============Erosi Tanpa Library=============
img_er = cv2.imread('/content/drive/MyDrive/PCVK/Images/j.png',0)

#Acquire size of the image
m, n = img_er.shape
# shape itu nantinya akan menyimpan nilai m dan n untuk menyimpan nilai lebar dan tinggi

#Show the image
plt.imshow(img_er, cmap="gray")

# Define the structuring element
# k= 11,15,45 -Different sizes of the structuring element
# k sebagai structuring elemen
k = 5
SE = np.ones((k,k), dtype=np.uint8)
constant = (k-1)//2
# maksud dari //2 mengambil nilai berupa int

#Define new image
imgErode = np.zeros((m,n), dtype=np.uint8)

#Erosion without using inbuilt cv2 function for morphology
for i in range(constant, m-constant): #constant adalah tipe variable yang nilainya tidak dapat diubah
  for j in range(constant,n-constant):
    temp = img_er[i-constant:i+constant+1, j-constant:j+constant+1] #temp digunakan untuk menampung atau menyimpan data untuk sementara
    product = temp * SE
    imgErode[i,j] = np.min(product)

plt.subplot(131),plt.imshow(img_er, cmap="gray"),plt.title('Citra Awal')
plt.subplot(132),plt.imshow(imgErode, cmap="gray"),plt.title('Citra Hasil Erosi')
```

```python
# 5. Erosi tanpa library
img2 = cv2.imread('/content/drive/MyDrive/PCVK/Images/j.png')

kernel = np.ones((5,5), np.uint8)
erosi = cv2.erode(img2, kernel, iterations = 1)

plt.subplot(121), plt.imshow(img2), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(erosi), plt.title('Citra Hasil Erosi')
```

```python
# 6. Erosi dengan library
kernel = np.ones((5,5), np.uint8)
erosi = cv2.morphologyEx(img2, cv2.MORPH_ERODE, kernel)

plt.subplot(121), plt.imshow(img2), plt.title('Citra Awal')
plt.subplot(122), plt.imshow(erosi), plt.title('Citra Hasil Erosi')
```
